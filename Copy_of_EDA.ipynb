**Filling Null Values with Most Frequent Values**

```python
df2=pd.read_csv("/content/heart copy.csv")
```

```python
df2.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='most_frequent')
df2['Age'] = imputer.fit_transform(df2[['Age']])
```

**Checking for Null in each column after filling**

```python
df2.isnull().sum()
```

**Replacing Null values with Default Values**

```python
df3=pd.read_csv("/content/heart copy.csv")
```

```python
df3.isnull().sum()
```

```python
default_value = 50
df3['Age'] = df3['Age'].fillna(default_value)
```

**Checking for Null in each column after filling**

```python
df3.isnull().sum()
```

**Filling Null values with Min Values**

```python
df4=pd.read_csv("/content/heart copy.csv")
```

```python
df4.isnull().sum()
```

```python
df4 = df4.fillna(value= df4['Age'].min())
```

**Checking for Null in each column after filling**

```python
df4.isnull().sum()
```

**Filling Null Values with Max Values**

```python
df5=pd.read_csv("/content/heart copy.csv")
```
